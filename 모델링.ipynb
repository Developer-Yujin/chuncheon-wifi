```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# 데이터 로딩
df_wifi = pd.read_csv('./와이파이.csv', encoding='euc-kr')
df_elder_center = pd.read_csv('./춘천시 경로당.csv', encoding='euc-kr')

# 선형 회귀 모델 생성 및 학습
X = df_wifi[['val']]
y_traffic = df_wifi['트래픽_y']
y_user = df_wifi['사용자_y']

reg = LinearRegression()
reg.fit(X, y_traffic)

# 경로당 데이터에 대한 예측 (트래픽)
df_elder_center['트래픽'] = reg.predict(df_elder_center[['val']])

# 모델 재사용하여 사용자 예측
reg.fit(X, y_user)
df_elder_center['사용자'] = reg.predict(df_elder_center[['val']])
df_elder_center.sort_values('트래픽',ascending=False)
```
