```python
#사용자
q3_users = df_사용자['사용자'].quantile(0.75)
high_users_aps = df_사용자[df_사용자['사용자'] >= q3_users]

#트래픽(Bytes)
q3_bytes = df_트래픽['트래픽'].quantile(0.75)
high_bytes_aps = df_트래픽[df_트래픽['트래픽'] >= q3_bytes]

q1_users = df_사용자['사용자'].quantile(0.25)
low_users_aps = df_사용자[df_사용자['사용자'] <= q1_users]
q1_bytes = df_트래픽['트래픽'].quantile(0.25)
low_bytes_aps = df_트래픽[df_트래픽['트래픽'] <= q1_bytes]
```

# a. 사용자수 많고 트래픽양 많음

```python
df1 = pd.merge(high_users_aps, high_bytes_aps, on='name')
```

```python
df1
```

```python
import folium
from folium.plugins import MarkerCluster

# 지도 중심을 데이터의 평균 위치로 설정
map_center = [df1['lat_x'].mean(), df1['lng_x'].mean()]

# 지도 객체 생성
map_ = folium.Map(location=map_center, zoom_start=12)

# 마커 클러스터 객체 생성
marker_cluster = MarkerCluster().add_to(map_)

# 데이터프레임을 순회하며 마커 추가
for idx, row in df1.iterrows():
    folium.Marker(
        location=[row['lat_x'], row['lng_x']],  # 위도, 경도
        popup=f"{row['name']}<br>사용자: {row['사용자_x']}<br>트래픽: {row['트래픽_x']}",  # 팝업에 표시될 정보
    ).add_to(marker_cluster)

# 지도 출력
map_
```

```python
count_gyoungrodang = df1['name'].str.contains('경로당').sum()
count_busstop = df1['name'].str.contains('버스정류장').sum()
count_cctv = df1['name'].str.contains('CCTV').sum()
count_others = len(df1) - (count_gyoungrodang + count_busstop + count_cctv)

# 결과 출력
counts = {
    '경로당': count_gyoungrodang,
    '버스정류장': count_busstop,
    'CCTV': count_cctv,
    '그 외': count_others
}

# 데이터 시각화 (파이 차트)
labels = list(counts.keys())
sizes = list(counts.values())
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Wi-Fi 재배치 기준별 데이터 분포")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# 데이터 로드 (예시)
# 데이터프레임이 'data'라는 이름으로 되어 있다고 가정합니다.
# '사용자_y'와 'val' 열을 추출
X = level_d_sorted[['사용자_y']].values.reshape(-1, 1)  # 사용자 데이터를 X로 사용
y = level_d_sorted['val'].values  # val 데이터를 y로 사용

# 1. 상관계수 확인
correlation = np.corrcoef(level_d_sorted['사용자_y'], level_d_sorted['val'])[0, 1]
print(f'사용자와 val 간의 상관계수: {correlation}')

# 2. 선형 회귀 모델 적용
model = LinearRegression()
model.fit(X, y)

# 예측값 계산
y_pred = model.predict(X)

# 3. R^2 값 계산 (결정 계수)
r2 = r2_score(y, y_pred)
print(f'R^2 (결정 계수): {r2}')

# 4. 산점도와 선형 회귀선 시각화
plt.scatter(level_d_sorted['사용자_y'], level_d_sorted['val'], color='blue', label='Actual Data')
plt.plot(level_d_sorted['사용자_y'], y_pred, color='red', label='Regression Line')
plt.xlabel('사용자 수')
plt.ylabel('val 값')
plt.title(f'사용자와 val 간의 선형 회귀 (상관계수: {correlation:.2f}, R^2: {r2:.2f})')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df7 = pd.read_csv("/content/sample_data/cor_feature.csv", encoding='cp949')
```

```python
df7
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
import numpy as np
df7['facilities'] = df7[['caf?','resturant','accom']].sum(axis=1)

# 2. train, bus를 합쳐서 새로운 열 생성
df7['transport'] = df7[['train', 'bus']].sum(axis=1)
variance = df7[['facilities', 'transport', 'culture', 'public']].var()

scaler = MinMaxScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(df7[['facilities', 'transport', 'culture', 'public']]),
                           columns=['facilities', 'transport', 'culture', 'public'])

# 2. 주성분 분석 (PCA)
pca = PCA(n_components=4)
pca.fit(scaled_data)
pca_weights = pca.explained_variance_ratio_

# 주성분 분석 결과를 바탕으로 가중치 부여
weights = pca_weights / pca_weights.sum()

# 각 변수별 가중치
weight_dict = {
    'facilities': weights[0],
    'transport': weights[1],
    'culture': weights[2],
    'public': weights[3]
}

# 가중치 적용한 결과 계산 (weighted sum)
df7['weighted_sum'] = (
    df7['facilities'] * weight_dict['facilities'] +
    df7['transport'] * weight_dict['transport'] +
    df7['culture'] * weight_dict['culture'] +
    df7['public'] * weight_dict['public']
)
```
